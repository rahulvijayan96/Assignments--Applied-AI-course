```python
dataset = pd.read_csv('5_a.csv')
dataset.head()
```

```python

dataset.loc[(dataset['proba']>0.5,'proba')] = 1
dataset.loc[dataset['proba']<0.5,'proba'] = 0

df = dataset.rename(columns={'proba':'y_pred'})
df.head()
```

CONFUSION MATRIX

```python
def confusion_matrix(actual,predicted):
    TP=0;TN=0;FP=0;FN=0
    for i in range(len(predicted)):
        if (predicted[i] == actual[i] == 1) :
            TP+=1
        if (predicted[i] == actual[i] == 0) :
            TN+=1
        if (predicted[i] - actual[i] == 1) :
            FP+=1
        if (actual[i] - predicted[i] == 1) :
            FN+=1
    return TP,TN,FP,FN 

TP,TN,FP,FN=confusion_matrix(df['y'],df['y_pred'])
confusion_mat = [[TN, FP],
 [FN, TP]]

print("confusion matrix : ",confusion_mat)
```

F1 SCORE

```python
precision = TP/(TP+FP)
recall = TP/(TP+FN)

F1_score = 2*precision*recall / (precision + recall)
print('F1_score = ',F1_score)
```

ACCURACY SCORE

```python
# acc = no. of correctly classified pts / total no. of pts

accuracy_score = (TP+TN)/(TP+FP+TN+FN)
print("Accuracy Score : ",accuracy_score)

dataset['proba'][0]
```

ROC CURVE

```python
##   https://stackoverflow.com/questions/61321778/how-to-calculate-tpr-and-fpr-in-python-without-using-sklearn


dataset = pd.read_csv('5_a.csv')
dataset = dataset.sort_values(by = ['proba'],ascending = False)   # SORT VALUES IN DESCENDING ORDER OF PROBABILITY SCORES
dataset = dataset.reset_index(drop=True)                          # RESET INDICES

def predicted_at_threshold(threshold, actual, predicted):
    p = np.where(predicted >= threshold, 1, 0) 
    tp, tn, fp, fn = confusion_matrix(actual,p)
    tpr = (tp)/(fn+tp)
    fpr = (fp)/(tn+fp)
    return fpr, tpr
def roc_curve(actual,predicted,thresholds):    
    tpr_2 = []
    fpr_2 = []
    for i in thresholds:        
        fpr_t, tpr_t = predicted_at_threshold(i, actual , predicted)
        tpr_2.append(tpr_t)
        fpr_2.append(fpr_t)
    return fpr_2, tpr_2

FPR,TPR = roc_curve(dataset['y'],dataset['proba'],dataset['proba'])
```

AUC SCORE

```python
pAUC = np.trapz(TPR, FPR)

print("AUC SCORE : ",pAUC)
```

```python

def predicted_at_threshold_c(threshold, actual, predicted):
    p = np.where(predicted >= threshold, 1, 0)
    tp, tn, fp, fn = confusion_matrix(actual,p)
    return fn,fp

def roc_curve_best_threshold(actual,predicted,thresholds):    
    A={}
    for i in thresholds:
        fn_t,fp_t = predicted_at_threshold_c(i, actual, predicted)
        A.update({(500*fn_t + 100*fp_t) : (i)})        
    return A

A = roc_curve_best_threshold(dataset_3['y'],dataset_3['prob'],dataset_3['prob'])
```

```python
dataset_4 = pd.read_csv('5_d.csv')
dataset_4.head()
```

MEAN SQUARE ERROR

```python
# mean square error = 1/n (summation of 1->n)((y-y_pred)^2)
sq_err = 0
for i in range(len(dataset_4)):
    sq_err +=  ( dataset_4['y'][i] - dataset_4['pred'][i] )**2
mean_sq_err =  sq_err / len(dataset_4)
print("MEAN-SQUARE ERROR : ",mean_sq_err)
```

MAPE

```python
# from video, simplified eqn of M-MAPE = (sum of all absolute val of predcited values)  /  (sum of all actual values)
```

```python
abs_val_sum = 0
act_val_sum = 0

abs_val = dataset_4['pred'].abs()      # absolute value of all items of 'y_pred'
abs_val_sum = abs_val.sum()            # sum of all absolute val of predcited values
act_val_sum =  dataset_4['y'].sum()    # sum of all actual values

MAPE = abs_val_sum / act_val_sum 
print("MAPE : ",MAPE)
```

R^2 ERROR

```python
# total sum of squares - SS_tot
# residual sum of squares - SS_res
    
y_bar = dataset_4['y'].mean()
y_bar
SS_tot=0
SS_res = 0
for j in range(len(dataset_4)):
    SS_tot += (dataset_4['y'][j] - y_bar )**2
    SS_res += (dataset_4['y'][j] - dataset_4['pred'][j] )**2

R_2_error  = 1 - (SS_res/SS_tot)
print("R^2_ERROR is ",R_2_error)
```
